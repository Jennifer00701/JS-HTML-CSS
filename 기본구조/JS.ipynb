# 비교연산

JS는 주어진 두 항을 비교할 수 있는 비교 연산자를 제공함
    2가지 유형을 비교할 수 있다.
* 대소 비교 = 크냐 작냐
* 등가 비교 = 같냐 다르냐 
    무언갈 비교하는 작업 또한 연산의 일종이라 할 수 있음

비교 연산식은 언제나 boolean 데이터를 반환함
위 2가지 유형을 비교 모두 하나의 질문이며 대답이 참, 거짓이다 
    예시 : 
        대소 비교 = 3이 5보다 작냐 = true
        등가 비교 = 6과 6이 다르냐 = false

# 비교 연산자 
    > 크다, < 작다, >= 크거나 같다, <=작거나 크다
    == 같다, != 같지 않다, === 완전히 같다, !==완전히 같지 않다
            등호(이퀄사인,=)하나는 대입 연산자이므로 기호 개수를 잘 봐가며 사용 ! ***  

= 대입 연산자기 때문에 비교 시 두개 이상 써야 함 
== 는 추상적 같은 비교로 자료형이 다르더라도 같다고 판단할 수 있다.
=== 는 엄격한 같은 비교로 자료형 데이터가 모두 일치해야만 같다고 판단
    console.log('1'==1) true
    console.log('1'===1) false
